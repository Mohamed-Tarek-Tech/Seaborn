```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('Mall_Customers.csv')
df
```

### regplot()

```python
sns.regplot(x = df['Age'], y = df['Annual Income (k$)'])
```

```python
sns.regplot(x = df['Age'], y = df['Spending Score (1-100)'])
```

```python
sns.regplot(x = df['Age'], y = df['Spending Score (1-100)'], label = 'Male')
sns.regplot(x = df['Age'], y = df['Spending Score (1-100)'], label = 'Female')
plt.legend()
# Note: replot() doesn't have 'hue' argument for categorical data
```

### lmplot()

```python
# Note: "lmplot()" has (x,y) arguments, and they expect to take a column names, not values.
# sns.lmplot(x = df['Age'], y = df['Spending Score (1-100)'], data = df)   If we write it, it will get and error.
# "lmplot()" is similar to "regplot", but it's more powerful.

sns.lmplot(x = 'Age', y = 'Spending Score (1-100)', data = df)
```

### lmplot faceting:

1. Organizing data by colors --> hue
2. Organizing data by columns --> col
3. Organizing data by row --> row 

```python
# 1.
sns.lmplot(x = 'Age', y = 'Spending Score (1-100)',hue = 'Genre' ,data = df)
```

```python
# 2.
sns.lmplot(x = 'Age', y = 'Spending Score (1-100)',col = 'Genre' ,data = df)
```

```python
# 3.
sns.lmplot(x = 'Age', y = 'Spending Score (1-100)',row = 'Genre' ,data = df)
```
